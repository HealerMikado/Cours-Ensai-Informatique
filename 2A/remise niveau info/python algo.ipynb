#### Les instructions break et continue

Par moment vous allez avoir besoin de sauter ou de sortir pr√©matur√©ment de vos boucles, pour cela vous avez les instructions `break` (sortir de la structure) et `continue` (sauter une √©tape).

> √Ä titre personnel je d√©conseille leur utilisation. En effet si dans de rare cas elles peuvent √™tre utiles, dans d'autre elles rajoutent de la complexit√© cognitive au programme en le rendant plus dur √† comprendre et analyser. Et globalement je pense qu'utiliser un break implique d'avoir pris une boucle for alors qu'il fallait une while.

##### Break

Comme dit au dessus, un `break` permet de sortir d'une boucle. Dans le cas de boucles imbriqu√©es cela permet de sortir uniquement de la boucle dans laquelle vous vous trouvez.

Exemple :

````python
i = 0
print("Je rentre dans la boucle")
while True : #toujours vrai, donc on reste dans la boucle pour toujours (c'est une mauvaise pratique !!!!!)
    i+=1
    if i>10 :
        break
    print("working")
print("the end")
````

##### Continue

√Ä la diff√©rence de break, `continue` permet de terminer l'√©tape en cours, mais on reste dans la boucle

Exemple : 

````python
for c in " Twilight Sparkle ": # Exemple continue
    if (c == "i") :
        continue
    print (c, end="")
````

#### L'instruction `else` en fin de boucle

Il est possible de rajouter une instruction `else` √† la fin d'une boucle. Le code dans le bloc `else` sera ex√©cut√© uniquement si la boucle s'arr√™te naturellement (donc sans un `break` ou une erreur). Voil√† √† quoi √ßa ressemble :

````python
for c in " Twilight Sparkle ": # Exemple continue
    if (c == "i") :
        continue
    print (c, end="")
else :
    print("On est rentr√© dans le else")
````

````python
for c in " Twilight Sparkle ": # Exemple continue
    if (c == "i") :
        break
    print (c, end="")
else :
    print("On est rentr√© dans le else")
````

## Les modules

Les modules permettent d'√©clater votre code entre diff√©rents fichiers pour qu'il soit plus lisible et maintenable. Un peu comme les fonctions, chaque module va contenir du code relatif √† un th√®me ou un type de manipulation. Plus factuellement, un module est un fichier *.py* qui peut contenir des variables, des fonctions, des classes (cf python orient√© objet). Ce fichier peut √™tre appel√© par votre programme principal ou un autre module. Python dispose de bases de plusieurs modules importables mais rapidement pour des taches sp√©cifique (machine larning par exemple), vous aller devoir t√©l√©charger des module suppl√©mentaire.

Pour r√©aliser un import deux syntaxes :

- ```python
  import modname [as alias]
  ```

Permet d'importer le module souhait√© en entier et de l'utiliser en pr√©fixant l'objet que l'on souhaite utiliser par le nom du module. Par exemple

````python
import math
math.sin(1)
````

Il est √©galement possible de donner un alias au module et d'utiliser l'alias par la suite

````python
import math as m
m.sin(1)
````

- ```python
  from modname import attname [as alias]
  ```
 Permet d'importer seulement certains attributs (variables, fonctions, classes) d'un module.

````python
from math import sin
sin(1)
````

Il est √©galement possible de donner un alias l'attribut import√©

````python
from math import sin as sinus
sinus(1)
````

Il est aussi possible d'importer tout un module avec une syntaxe from

````python
from math import *
sin(1)
````

> **import ... vs from ... import ...**
>
> Comme vous avez pu le constater avec la syntaxe from ... import ... pas besoin de r√©p√©ter le nom du module que l'on utilise lors de l'invocation d'un attribut. Si cela peut sembler une bonne chose ce n'est pas vrai. En effet cela pose deux probl√®mes
> - On ne sait pas si l'attribut est interne ou externe √† notre script
> - Un attribut externe peut rentrer en concurence avec un attribut interne (surtout avec un from module import \*)
> 
>N√©anmoins, la syntaxe from ... import ... permet d'importer que des attributs sp√©cifiques. Donc je vous conseille d'utiliser from ... import ... avec une liste d'attributs (jamais \*) quand vous √™tes sur d'une absence de conflit, et sinon la syntaxe import.
