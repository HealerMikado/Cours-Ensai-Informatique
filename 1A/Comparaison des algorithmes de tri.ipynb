# Comparaison des algorithmes de tri

## Tri par insertion

Fonction utilitaire de création de de liste aléatoire

```python
import random

def randomList(lenght) :
    mylist = []

    for i in range(0,lenght):
        x = random.randint(1,1000000)
        mylist.append(x)
    return mylist
```

Notre aglo de tri par insertion

```python
def insertionSort (listToSort) :
    i = 1
    currentElement = None
    while i<len(listToSort) :
        j = i-1
        currentElement = listToSort[i]
        while j>=0 and listToSort[j] > currentElement :
            listToSort[j+1] = listToSort[j]
            j = j - 1
        listToSort[j+1] = currentElement
        i +=1
    return listToSort
```

Maintenant on va le faire tourner avec des données de plus en plus grande et voir l'évolution des temps

```python
import time

timesUnsortedData = []
for i in range(300) :
    listToSort = randomList(10*i)
    start = time.time()
    insertionSort(listToSort)
    end = time.time()
    timesUnsortedData.append(end-start)
    print(i, end=",")
```

On applique l'ago sur des données déjà sortée

```python
timesSortedData = []
for i in range(300) :
    start = time.time()
    insertionSort([j for j in range(10*i)])
    end = time.time()
    timesSortedData.append(end-start)
    print(i, end=",")
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

plt.plot(timesUnsortedData, color="red")
plt.plot(timesSortedData, color="blue")
plt.ylabel('Temps de tri')
plt.show()
```

Le temps de tri monte rapidement quand on a des données alétoire (O(n²)). Pour des données déjà triée cela n'a pas l'air de prendre plus de temps (mais c'est un O(n)) quand même.

## Selection sort

```python
def selectionSort(listToSort) :
    # Traverse through all array elements 
    for i in range(len(listToSort)): 

        # Find the minimum element in remaining  
        # unsorted array 
        min_idx = i 
        for j in range(i+1, len(listToSort)): 
            if listToSort[min_idx] > listToSort[j]: 
                min_idx = j 

        # Swap the found minimum element with  
        # the first element
        temp = listToSort[i]
        listToSort[i] = listToSort[min_idx]
        listToSort[min_idx] = temp
```

```python
timesUnsortedDataSelectionSort = []
for i in range(300) :
    start = time.time()
    selectionSort(listToSort)
    end = time.time()
    timesUnsortedDataSelectionSort.append(end-start)
    print(i, end=",")
```

```python
timesSortedDataSelectionSort = []
for i in range(300) :
    start = time.time()
    selectionSort([j for j in range(10*i)])
    end = time.time()
    timesSortedDataSelectionSort.append(end-start)
    print(i, end=",")
```

```python
%matplotlib notebook

plt.plot(timesUnsortedDataSelectionSort, color="red")
plt.plot(timesSortedDataSelectionSort, color="blue")
plt.ylabel('Temps de tri')
plt.show()
```

Preuve que dans certain cas avoir des données déjà triée n'apporte rien
